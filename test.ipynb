```python
import json
```

```python
my = {"catalog": []}
with open("news.json", 'r', encoding="utf-8") as f:
    for line in f:
        my["catalog"].append(json.loads(line))
```

```python
len(my["catalog"])
```

```python
print(json.dumps(my["catalog"][0], ensure_ascii=False))
```

```python
with open("news_fix.json", "w", encoding="utf-8") as file:
    json.dump(my, file, ensure_ascii=False, indent=4)
```
